```python
import os, json, random
```

```python
sample = {}
with open("COCO/100/annotations/instances_default.json", "r") as f:
    sample = json.load(f)
```

```python
ALL = {}
ALL["licenses"] = sample["licenses"]
TALL["info"] = sample["info"]
ALL["categories"] = sample["categories"]
ALL["images"] = []
ALL["annotations"] = []
id_count = 1
annotation_id_count = 1
for directory in os.listdir("COCO"):
    partial = {}
    translation = {x+1:x+id_count for x in range(100)}
    if directory == "Archives":
        continue
    with open("COCO/%s/annotations/instances_default.json"%directory, "r") as f:
        partial = json.load(f)
    images = partial["images"]
    annotations = partial["annotations"]
    for image in images:
        image["id"] = id_count
        id_count += 1
        ALL["images"].append(image)
    for annotation in annotations:
        annotation["id"] = annotation_id_count
        annotation_id_count += 1
        annotation["image_id"] = translation[annotation["image_id"]]
        ALL["annotations"].append(annotation)
with open("instances_all.json", "w") as f:
    json.dump(ALL, f)
```

```python
all_data = set(os.listdir("all_knives"))
test_len = int(len(all_data)/7)
train_len = len(all_data) - test_len*2
train_list = random.sample(all_data, train_len)
train_data = set(train_list)
test_list = random.sample(all_data-train_data, test_len)
test_data = set(test_list)
val_data = all_data - train_data - test_data
val_list = list(val_data)
```

```python
with open("train.txt", "w") as f:
    f.writelines([x+"\n" if x!=train_list[-1] else x for x in train_list])
with open("test.txt", "w") as f:
    f.writelines([x+"\n" if x!=test_list[-1] else x for x in test_list])
with open("val.txt", "w") as f:
    f.writelines([x+"\n" if x!=val_list[-1] else x for x in val_list])
```

```python
TRAIN_ALL = {}
TRAIN_ALL["licenses"] = sample["licenses"]
TRAIN_ALL["info"] = sample["info"]
TRAIN_ALL["categories"] = sample["categories"]
TRAIN_ALL["images"] = []
TRAIN_ALL["annotations"] = []

TEST_ALL = {}
TEST_ALL["licenses"] = sample["licenses"]
TEST_ALL["info"] = sample["info"]
TEST_ALL["categories"] = sample["categories"]
TEST_ALL["images"] = []
TEST_ALL["annotations"] = []

VAL_ALL = {}
VAL_ALL["licenses"] = sample["licenses"]
VAL_ALL["info"] = sample["info"]
VAL_ALL["categories"] = sample["categories"]
VAL_ALL["images"] = []
VAL_ALL["annotations"] = []

ALL = {}
train_list = []
test_list = []
val_list = []
with open("instances_all.json", "r") as f:
    ALL = json.load(f)
with open("train.txt", "r") as f:
    train_list = [x.replace("\n", "") for x in f.readlines()]
with open("test.txt", "r") as f:
    test_list = [x.replace("\n", "") for x in f.readlines()]
with open("val.txt", "r") as f:
    val_list = [x.replace("\n", "") for x in f.readlines()]
```

```python
image_id = 1
anno_id = 1
translation = {}
for image in ALL["images"]:
    if image["file_name"] in test_list:
        translation[image["id"]] = image_id
        image["id"] = image_id
        TEST_ALL["images"].append(image)
        image_id += 1
for annotation in ALL["annotations"]:
    if ALL["images"][annotation["image_id"]-1]["file_name"] in test_list:
        annotation["image_id"] = translation[annotation["image_id"]]
        annotation["id"] = anno_id
        TEST_ALL["annotations"].append(annotation)
        anno_id += 1
with open("test.json", "w") as f:
    json.dump(TEST_ALL, f)
```

```python
image_id = 1
anno_id = 1
translation = {}
for image in ALL["images"]:
    if image["file_name"] in val_list:
        translation[image["id"]] = image_id
        image["id"] = image_id
        VAL_ALL["images"].append(image)
        image_id += 1
for annotation in ALL["annotations"]:
    if ALL["images"][annotation["image_id"]-1]["file_name"] in val_list:
        annotation["image_id"] = translation[annotation["image_id"]]
        annotation["id"] = anno_id
        VAL_ALL["annotations"].append(annotation)
        anno_id += 1
with open("val.json", "w") as f:
    json.dump(VAL_ALL, f)
```

```python
image_id = 1
anno_id = 1
translation = {}
for image in ALL["images"]:
    if image["file_name"] in train_list:
        translation[image["id"]] = image_id
        image["id"] = image_id
        TRAIN_ALL["images"].append(image)
        image_id += 1
for annotation in ALL["annotations"]:
    if ALL["images"][annotation["image_id"]-1]["file_name"] in train_list:
        annotation["image_id"] = translation[annotation["image_id"]]
        annotation["id"] = anno_id
        TRAIN_ALL["annotations"].append(annotation)
        anno_id += 1
with open("train.json", "w") as f:
    json.dump(TRAIN_ALL, f)
```
